```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
df = pd.read_csv('sample_smiles.csv.csv')
```

```python
print(df.head())
```

```python
df['Molecule'] = df['Smiles'].apply(lambda x: Chem.MolFromSmiles(x))
mols = df['Molecule'].tolist()
names = df['ChemBL ID'].tolist()
img = Draw.MolsToGridImage(mols, molsPerRow=3, legends=names, subImgSize=(200,200))
from IPython.display import display
display(img)
```
